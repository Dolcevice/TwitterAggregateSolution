```python
# Remove all trailing spaces and punctuation
clean_df.columns.str.rstrip('.!? \n\t')
# Remove all newlines
clean_df.replace('\n', '', regex=True, inplace=True)
# Remove all twitter media links
clean_df.replace('tco.{10}', '', regex=True, inplace=True)
clean_df.replace('tco', '', regex=True, inplace=True)
clean_df.replace('https:.{13}', '', regex=True, inplace=True)
clean_df.replace('https:', '', regex=True, inplace=True)
# Remove all extra punctuations
clean_df.replace('[,\.!?/|_]', '', regex=True, inplace=True)
# Remove retweet marks
clean_df.replace(r'RT.*:', '', regex=True, inplace=True)
clean_df.replace(r'\b\w{1,2}\b', '', regex=True, inplace=True)
clean_df.replace('@.* \s', '', regex=True, inplace=True)


# Show sorted values
clean_df.sort_values(by = ['retweet_count'], ascending=False)
```

```python
# Convert the dataframe into text
str_dump = clean_df['text'].to_string()
print(str_dump)
# os.remove('tweet_dump.json')
```

```python
# Import sklearn
from sklearn.feature_extraction.text import CountVectorizer as CV
from sklearn.linear_model import LogisticRegression as LogicReg
from sklearn.metrics import accuracy_score as acc
from sklearn.model_selection import train_test_split as train
```

```python
# Get training Data
# From https://github.com/aaronkub/machine-learning-examples/blob/master/imdb-sentiment-analysis/movie_data.tar.gz
# Create Array
reviews_train = []
for line in open(r"C:\Users\Dolcevice\Documents\GitHub\TwitterAggregateSolution\common\Jupyter\Train_Data\aclImdb\movie_data\full_train.txt", 'r',encoding='utf-8'):
    reviews_train.append(line.strip())
reviews_test = []
for line in open(r"C:\Users\Dolcevice\Documents\GitHub\TwitterAggregateSolution\common\Jupyter\Train_Data\aclImdb\movie_data\full_test.txt", 'r',encoding='utf-8'):
    reviews_test.append(line.strip())
```

```python
# Clean training Data
REPLACE_NO_SPACE = re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

reviews_train_clean = [REPLACE_NO_SPACE.sub("", line.lower()) for line in reviews_train]
reviews_test_clean = [REPLACE_NO_SPACE.sub("", line.lower()) for line in reviews_test]
```

```python
# Vectorize
cv = CV(binary = True)
cv.fit(reviews_train_clean)
X = cv.transform(reviews_train_clean)
X_test = cv.transform(reviews_test_clean)
```

```python
# Classifier
target = [1 if i < 12500 else 0 for i in range(25000)]
X_train, X_val, y_train, y_val = train(X, target, train_size = 0.75)

# Use logistic regression on the training data
for c in [0.01, 0.05, 0.25, 0.5, 1]:
    lr = LogicReg(C=c)
    lr.fit(X_train, y_train)
    print ("Accuracy for C=%s: %s" % (c, acc(y_val, lr.predict(X_val))))
```

```python
# Final Training Model

final_model = LogicReg(C=0.05, n_jobs=-1)
final_model.fit(X, target)
print ("Final Accuracy: %s" % acc(target, final_model.predict(X_test)))
```

```python
# Save final model
pickle.dump(final_model, open('final_model.sav', 'wb'))
```
